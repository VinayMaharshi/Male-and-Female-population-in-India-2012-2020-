```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv(r"D:\Data Science\data sets\Female population in India.csv")
df
```

```python
df.isnull()
```

```python
newdf=df.iloc[5:14].dropna()
newdf
```

```python
newdf.dtypes
```

```python
newdf['Time']=newdf['Time'].astype(int)
newdf['Value']=newdf['Value'].astype(int)
newdf.dtypes
```

```python
newdf
```

```python
new=pd.read_csv(r"D:\Data Science\data sets\Male population in India.csv")
new
```

```python
new.isnull()
```

```python
male=new.iloc[5:14].dropna()
male
```

```python
male.dtypes
```

```python
male['Value']=male['Value'].astype(int)
male['Time']=male['Time'].astype(int)
male.dtypes
```

```python
male
```

```python
plt.figure(figsize=(12,8)) #setting the figure size




plt.plot(newdf['Time'],newdf['Value'],marker='o') #plotting graph 1
plt.plot(male['Time'],male['Value'],marker='o') #plotting graph 2

#changing the values of ticks
plt.xticks(fontsize=10,weight='bold')
plt.yticks(fontsize=10,weight='bold')

#setting title and labels.
plt.title('Population graph (Male and Female) in India.'
          ,fontname='Times New Roman',fontsize=20,pad=15)
plt.xlabel('Year',fontname='Times New Roman',fontsize=20,labelpad=15)
plt.ylabel('Number in Crores',fontname='Times New Roman'
           ,fontsize=20,labelpad=15)

#adding legend to represent the graphs.
plt.legend(['Female Population','Male Population'],fontsize=15)

#use the grid function for better understanding of the graph.
plt.grid()
plt.show()
```
